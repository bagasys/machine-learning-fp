```python
import pandas as pd
df = pd.read_csv("cc.csv")
inputs = df.drop("CUST_ID", axis='columns')
inputs = inputs[inputs.columns[0:17]]
```

```python
import numpy as np
inputs[np.isnan(inputs)] = 0
```

```python
from sklearn.cluster import KMeans
km = KMeans(
    n_clusters=2, init='random',
    n_init=100, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit(inputs)
# print(y_km.cluster_centers_)
# print(y_km.labels_)
new = inputs.to_numpy()
new
plt.scatter(new[:,5],new[:,6], c=y_km.labels_, cmap='rainbow')
plt.scatter(y_km.cluster_centers_[:,0] ,y_km.cluster_centers_[:,1], color='black')

# inputs.iloc[:1,:]
```
