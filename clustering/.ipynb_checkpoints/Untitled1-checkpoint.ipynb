```python
import pandas as pd
df = pd.read_csv("cc.csv")
inputs = df.drop("CUST_ID", axis='columns')
inputs = inputs[inputs.columns[0:17]]
```

```python
import numpy as np
inputs[np.isnan(inputs)] = 0
print(inputs)
```

```python
from sklearn.cluster import KMeans
km = KMeans(
    n_clusters=2, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(inputs)
type(inputs)
# inputs.iloc[:1,:]
```

```python
inputs['cl'] = y_km
inputs.plot.scatter(x ='PURCHASES', y='BALANCE',  c='cl', colormap='gist_rainbow')

```

```python
from sklearn.cluster import KMeans

n = 1000
d = pd.DataFrame({
        'x': np.random.randint(0,100,n),
        'y': np.random.randint(0,100,n),
    })

m = KMeans(5)
m.fit(d)

d['cl'] = m.labels_
d.plot.scatter('x', 'y', c='cl', colormap='gist_rainbow')
```
